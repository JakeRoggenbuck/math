```julia
f(x) = 3x^3 / 2x^4

u(x) = 3x^3
v(x) = 2x^4

# d/dx [u(x) / v(x)] = ( v(x) up(x) - u(x) vp(x) ) / v(x)^2
# up(x) and vp(x) are the derivatives of u(x) and v(x) respectively 
up(x) = 9x^2
vp(x) = 8x^3

fp(x) = ( v(x)up(x) - u(x)vp(x) ) / v(x)^2
```

```julia
using Plots;
r = range(-20, stop=20, length=180)
plot(r, [f, fp], xlim=(-20, 20), ylim=(-10, 10), label = ["f(x)" "fp(x)"])
```
