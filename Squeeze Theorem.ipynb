```julia
using Plots
X = -6:6;
```

```julia
f(x) = -(1/3)x^3 + x^2 -(7/3)
h(x) = cos((pi/2)x)
# g(x) = ?
# lets suppose f(x) =< g(x) =< h(x) when x is close to 2
```

```julia
plot(X, [f, h])
```

```julia
# Let's take the limit of f(x) as x approaches 2

# lim f(x) = lim(-(1/3)x^3 + x^2 -(7/3))
# x->2       x->2

# lim f(x) = -(1/3)(2)^3 + (2)^2 -(7/3)
# x->2

# lim f(x) = -1
# x->2

lim_fx_to_2 = -(1/3) * 2^3 + 2^2 -(7/3)
```

```julia
# Now let's take the limit of h(x) as x approaches 2

# lim h(x) = lim(cos((pi/2)x))
# x->2       x->2

# lim h(x) = cos((pi/2) * 2) = cos(pi)
# x->2

lim_hx_to_2 = cos(pi)
```
