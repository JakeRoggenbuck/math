```julia
# A simple function for use to find the derivative of
f(x) = 2x^3
```

```julia
h = .00001

# The value we pass into each function
c = 4

# The actual output of the function
j = f(c)

# The approximation of it's slope for x = c
k = ( f(c + h) - f(c) ) / h

println("The output of the function when x is ", c, " is ", j);
println("The approximated output of the derivative when x is ", c, " is ", k);
```

```julia
# If you take the derivative of f(x) without approximation, you get the function fp(x)
fp(x) = 6x^2

l = fp(c)
println("The output of the derivative when x is ", c, " is ", l);
```

```julia
println("The approximated output is very close to the actual");
println("k - l = ", k - l);
```
