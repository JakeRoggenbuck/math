```julia
using Plots
```

```julia
random_line(x) = (x*2) * (rand((0.3, 0.8)) * rand((0.3, 0.4)))
```

```julia
x = 1.:100.
y = [random_line(a) for a in x]
```

```julia
plot(x, y)
```

```julia
N = length(x)

Î£_xy = sum([b*c for (b, c) in zip(x, y)])

Î£_x = sum(x)
Î£_y = sum(y)

Î£_x2 = sum([b^2 for b in x])
Î£_x_2 = sum(x)^2
```

```julia
m = (N * Î£_xy - (Î£_x * Î£_y)) / (N * Î£_x2 - Î£_x_2)
```

```julia
b = (Î£_y - m * Î£_x) / N
```

```julia
f(x) = m * x + b
```

```julia
plot(x, [y, [f(e) for e in x]])
```
