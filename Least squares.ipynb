```julia
using Plots
```

```julia
random_line(x) = (x*2) * (rand((0.3, 0.8)) * rand((0.3, 0.4)))
```

```julia
x = 1.:100.
y = [random_line(a) for a in x]
```

```julia
plot(x, y)
```

```julia
N = length(x)

Σ_xy = sum([b*c for (b, c) in zip(x, y)])

Σ_x = sum(x)
Σ_y = sum(y)

Σ_x2 = sum([b^2 for b in x])
Σ_x_2 = sum(x)^2
```

```julia
m = (N * Σ_xy - (Σ_x * Σ_y)) / (N * Σ_x2 - Σ_x_2)
```

```julia
b = (Σ_y - m * Σ_x) / N
```

```julia
f(x) = m * x + b
```

```julia
plot(x, [y, [f(e) for e in x]])
```
